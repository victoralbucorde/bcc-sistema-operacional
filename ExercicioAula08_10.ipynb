```python
import sys

class GerenciadorMemoria:

  def __init___(self, tamanho_total):
    self.memoria = [(True, tamanho_total)]

  def first_fit(self, tamanho):
    for i,(livre, tam_bloco) in enumerate(self.memoria):
      if livre and tam_bloco >= tamanho:
        if tam_bloco == tamanho:
          self.memoria[i] = [(False, tamanho)]
        else:
          self.memoria.insert(i+1,(True, tam_bloco - tamanho))
          self.memoria[i] = (False, tamanho)
        return i
    return None

  def best_fit(self, tamanho):
    temp_i = -1
    temp_tamanho = sys.maxsize - 1
    for i,(livre, tam_bloco) in enumerate(self.memoria):
      if livre and tam_bloco >= tamanho:
        if tam_bloco == tamanho:
          self.memoria[i] = [(False, tamanho)]
          return i
        elif tam_bloco < temp_tamanho:
          temp_tamanho = tam_bloco
          temp_i = i
    if i == -1:
      return None
    else:
      self.memoria.insert[i+1,(True, temp_tamanho - tamanho)]
      self.memoria[i] = [(False, tamanho)]

  def worst_fit(self, tamanho):
    temp_i = -1
    temp_tamanho = -1
    for i,(livre, tam_bloco) in enumerate(self.memoria):
      if livre and tam_bloco >= tamanho:
        if tam_bloco > temp_tamanho:
          temp_tamanho = tam_bloco
          temp_i = i
    if i == -1:
      return None
    else:
      self.memoria.insert[i+1,(True, temp_tamanho - tamanho)]
      self.memoria[i] = [(False, tamanho)]
```
